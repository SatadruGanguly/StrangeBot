```python
import pandas as pd
import nltk
import re
```

```python
data = pd.read_csv('Symptom-severity.csv')
```

```python
symptom_list = list(data['Symptom'][:132])
symptom_list
```

```python
nltk.download()
```

```python
stemmed_symptom_list = []
spaced_symptom_list = []
snow_stemmer = nltk.stem.snowball.SnowballStemmer(language='english')
```

```python
for symptom in symptom_list:
    words = symptom.split('_')
    new_word = ''
    for w in words:
        nw = snow_stemmer.stem(w)
        new_word = new_word + nw + ' '
    new_symptom_list.append(new_word.strip())
```

```python
new_symptom_list
```

```python
snow_stemmer.stem('shivering')
```

```python
new_dict = pd.DataFrame(new_symptom_list)
new_dict
```
